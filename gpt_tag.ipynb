```python
import os
import openai
import json
import csv
import pandas as pd

f = open("chatgptkey.txt", 'r')
line = f.readline()
openai.api_key = line


```

```python
df1 = pd.read_csv('drinkdata.csv')
frame=pd.DataFrame(df1)
df1.head(15)
```

```python


inputstr="주류 데이터를 검색하기 위한 태그를 @기호로 구분된 여러 개의 설명문들로부터 추출하고 싶어. 각 설명문마다 짧은 길이의 태그를 ,기호를 사용하여 10개 이내로 출력한 뒤 엔터를 출력하는 과정을 반복해줘."
outputs=[]
inputs=[]
msg=inputstr
inputcounts=0
for row in df1['텍스트 내용'] : #'average' 값을 얻어옴
  if(len(msg+row)>2000):
    inputs.append(msg)
    msg=inputstr
  msg= msg+"@"+row



print(len(inputs))
```

```python
import time
outputs=[]
for row in df1['텍스트 내용'] :
  description =  row # 원하는 설명문을 입력하세요

  prompt = f"주류 데이터를 검색하기 위한 태그를 각 설명문으로부터 추출하고 싶어. 다음 설명문에서 짧은 길이의 태그를 ,로 구분해서 10개 이내로 추출해줘.\n{description}"

  response = openai.Completion.create(
    engine="text-davinci-003",
    prompt=prompt,
    max_tokens=2000,  # 원하는 최대 토큰 수로 조정하세요
    temperature=0.3,  # 다양성을 조절할 수 있는 값, 0.2에서 1.0 사이의 값을 사용하세요
    stop=None  # 만약 특정 단어로 결과를 중단하고 싶지 않다면 None을 사용하세요
  )

  print(response.choices[0].text.strip())
  outputs.append(response.choices[0].text.strip())
  time.sleep(19)
```

```python
for i in range(3):
  outputs.append('')
```
