```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
from pandas import Series, DataFrame


names=[]
srcs=[]
for num in range(1,100):
  url = 'https://thesool.com/front/find/M000000082/list.do?searchType=2&searchKey=&searchKind=&levelType=&searchString=&productId=&pageIndex='+str(num)+'&categoryNm=&kind=#'
  header_info = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582'}
  page = requests.get(url, headers = header_info)
  page = page.text
  print(num)
  soup = BeautifulSoup(page, 'html.parser')
  infos=soup.find_all(attrs={'class':'info'})
  nameinfos=soup.find_all(attrs={'class':'name'})
  for i in nameinfos:
    names.append(i.text)
  imgs=soup.find_all(attrs={'class':'img'})
  for images in imgs:
    srcs.append("https://thesool.com/"+images.find("img").get("src"))
```

```python
imgtosrc=dict()
for ind in range(0,len(names)):
  imgtosrc[names[ind]]=srcs[ind]
```

```python
import csv
with open('imgsrcfile.csv','w') as f:
    w = csv.writer(f)
    w.writerow(imgtosrc.keys())
    w.writerow(imgtosrc.values())
```

```python
import time
from selenium.webdriver.common.by import By
for row in df1['술 이름']:
  if row in namelst:
    continue
  driver = webdriver.Chrome(options = specific_options)
  query=row
  alcid=imgtourldict[query]
  alcurl='http://211.37.148.214/api/alcohols/'+str(alcid)
  response=requests.get(alcurl,headers=headers)
  if(response.status_code==200):
    if(len(response.json()['images'])>0):
      continue
  driver.get(f'https://www.google.com/imghp')
  search_bar = driver.find_element(By.NAME,"q")
  search_bar.send_keys(query)
  search_bar.submit()
  last_height = driver.execute_script("return document.body.scrollHeight")
  new_height = 0

  while True :
    driver.execute_script("window.scrollBy(0,50)")
    time.sleep(PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height - last_height > 0 :
      last_height = new_height
      continue

    else :
      break

  img_elements = driver.find_elements(By.CSS_SELECTOR,".rg_i")[:1]

  for idx, img in enumerate(img_elements) :

    print(f"{query} : {idx+1}/{len(img_elements)} proceed...")

    img.click()
    time.sleep(PAUSE_TIME)

    img_element = driver.find_element(By.XPATH,'//*[@id="islrg"]/div[1]/div[1]/a[1]/div[1]/img')
    img_src = img_element.get_attribute('src')
    img_rst.append(img_src)
    namelst.append(query)



  driver.close()
print(img_rst)

```
