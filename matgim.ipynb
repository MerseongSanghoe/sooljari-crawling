```python

jsontexts=[]
for row in df1['텍스트 내용'] : #'average' 값을 얻어옴
    print(row)
    headerDict = {}
    headerDict.setdefault('x-auth-token', '4c79a20c-5d6f-4804-a3ea-18aec5012359')
    headerDict.setdefault('Content-Type', 'application/json')
    data={"document": row}
    requestData = requests.post('https://api.matgim.ai/54edkvw2hn/api-keyword', headers=headerDict, json=data)
    jsonData = None
    jsonData = requestData.json()
    jsontexts.append(jsonData['result']['keywords'])
df1['output'] = jsontexts
```
