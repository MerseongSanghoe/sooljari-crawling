```python
from konlpy.tag import Komoran
from collections import Counter

komoran = Komoran()

df1 = pd.read_csv('/content/drive/MyDrive/colab_github/sooljari-crawling/결과값.csv')

outputs=[]
for row in df1['텍스트 내용']:
  sentences_tag = []
  bucket_list = []
  try:
    pos_result = komoran.pos(row)
  except:
    pos_result=None
  sentences_tag.append(pos_result)
  #print(pos_result)

  if(pos_result):
    for my_sentence in sentences_tag:
      for word, tag in my_sentence:
        if tag in ['NN','NNP','VA','XR']:
          bucket_list.append(word)



  counts = Counter(bucket_list)
  print(counts.most_common())
  outputs.append(counts.most_common())

```

```python
print(len(outputs))
print(outputs)
df1['output2']=outputs
df1.to_csv('mid_output.csv')
```
