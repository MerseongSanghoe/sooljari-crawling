```python
from konlpy.tag import Okt

okt = Okt()
df1 = pd.read_csv('/content/drive/MyDrive/colab_github/sooljari-crawling/mid_output.csv')
okt.pos("형태소 분석을 테스트하고 결과를 출력해 보겠습니다")
output=[]
for row in df1['output']:
  #[a,b]로 이루어진 문자열
  refined=[]
  if(type(row)==type('output')):
   row=row.replace("[","").replace("]","")
   #print(row)
   lst=row.split(',')
   for text in lst:
    text=text.replace("'","").replace(" ","")
    if text.isdigit()==False:
      refined.append(text.replace("'",""))
  refined2=[]
  for re in refined:
    mixedstr=""
    for word, tag in okt.pos(re):
      if tag!='Foreign':
        mixedstr=mixedstr+word
    refined2.append(mixedstr)
  output.append(refined2)
df1['refinedOutput']=output

```
